```python
import requests
url="https://en.wikipedia.org/wiki/Main_Page"
headers={
    "User-Agent":"Mozilla/5.0"  # This makes the request look like it is coming from a real web browser instead of a script.
}
page=requests.get(url,headers=headers)
print(page.status_code)
print(page.text)

# we sucessfully hit the url and sucess code 200 confirming that
```

# Parsing HTML Using BeautifulSoup

 1.  BeautifulSoup cannot fetch pages by itself, it only parses HTML content
 2.  soup.prettify(): prints the HTML in a readable, structured format.

```python
!pip install bs4
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

# Extracting Data with BeautifulSoup

```python
print(soup.find_all('p'))  # Find all the paragraph tags
print("\n\n")  # print two new line fro readiability
print(soup.find_all('p')[0].get_text())  # prints the text of the first paragraph on the page
```

```python
# takes help from wikipedia inspect and than get the tags info 
container=soup.find(id="mp-left")
items=container.find_all(class_="mp-h2")
result=items[0]
print(result.prettify())
```

```python
print(result.get_text())
```
